3. Si yo fuera el líder del proyecto y quisiera explicar por qué una persona tiene obesidad, elegiría como variable de respuesta el "NivelDeObesidad". Esta es una variable cualitativa. Usaría esta variable para investigar si los otros factores están relacionados con los distintos niveles de obesidad.

La linea de codigo uitiliza la funcion value_counts() para contar y luego imprimir el nivel de obesidad que tienen las personas del estudio. 

```python
print(data.NivelDeObesidad.value_counts())
```

4. Para generar la grafica, primero se importa la libreria de matplorlib.pyplot y hacemos la grafica con la variable cuantitativa del peso exclusivamente en hombres en el eje x y en el eje y la variable de salida "NivelDeObesidad" solamente en hombres con un valor de transparencia (alpha) de 0.9.

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = data.Peso[data.Sexo == "Male"]
y = data.NivelDeObesidad[data.Sexo == "Male"]
alpha = 0.09
plt.scatter(x,y,alpha)
plt.show()
```
